```python
#Write a program to calculate the average of 5 subjects (Marks out of 100).

subject1 = float(input("Enter marks for Subject 1: "))
subject2 = float(input("Enter marks for Subject 2: "))
subject3 = float(input("Enter marks for Subject 3: "))
subject4 = float(input("Enter marks for Subject 4: "))
subject5 = float(input("Enter marks for Subject 5: "))

# Calculate total and average
total = subject1 + subject2 + subject3 + subject4 + subject5
average = total / 5

# Display results
print("Total Marks =", total)
print("Average Marks =", average)
```

```python
#Create a marks percentage calculator: take total marks obtained and maximum marks, then calculate percentage.


obtained_marks = float(input("Enter total marks obtained: "))
maximum_marks = float(input("Enter maximum marks: "))


percentage = (obtained_marks / maximum_marks) * 100


print("Percentage =", percentage , "%")
```

```python
#Calculate Simple Interest: (P * R * T) / 100 where P=10000, R=5.5, T=3.

P = 10000
R = 5.5
T = 3

simple_interest = (P * R * T) / 100

print("Simple Interest =", simple_interest)
```

```python
#Write a program that converts 5000 seconds into hours, minutes, and remaining seconds.

total_seconds = 5000

hours = total_seconds // 3600
remaining_seconds = total_seconds % 3600

minutes = remaining_seconds // 60
seconds = remaining_seconds % 60

print("Hours =", hours)
print("Minutes =", minutes)
print("Seconds =", seconds)
```
