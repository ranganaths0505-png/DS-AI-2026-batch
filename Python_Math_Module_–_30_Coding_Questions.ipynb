```python
# 28. Calculate compound interest using `math.pow()`.

# FV=PV(1+r)n

import math

principal = 1000
rate = 0.02
time = 5

amount = principal * math.pow(1 + rate, time)
compound_interest = amount - principal

print("Amount =", round(amount, 2))
print("Compound Interest =", round(compound_interest, 2))
```

```python
# 29. Find the area and circumference of a circle using `math.pi`.

# a=pir^2
# c=2pir

import math

radius = 6

area = math.pi * radius ** 2
circumference = 2 * math.pi * radius

print("Area =", area)
print("Circumference =", circumference)
```

```python
# 30. Create a scientific calculator that supports square root, factorial, logarithm, power, trigonometric functions, GCD, and LCM.

import math

print("Scientific Calculator")
print("1. Square Root")
print("2. Factorial")
print("3. Logarithm (base 10)")
print("4. Power")
print("5. Sine")
print("6. Cosine")
print("7. Tangent")
print("8. GCD")
print("9. LCM")

choice = int(input("Enter your choice (1-9): "))

if choice == 1:
    n = float(input("Enter a number: "))
    print("Square Root =", math.sqrt(n))

elif choice == 2:
    n = int(input("Enter a number: "))
    print("Factorial =", math.factorial(n))

elif choice == 3:
    n = float(input("Enter a number: "))
    print("Log10 =", math.log10(n))

elif choice == 4:
    base = float(input("Enter base: "))
    exp = float(input("Enter exponent: "))
    print("Power =", math.pow(base, exp))

elif choice == 5:
    angle = float(input("Enter angle in degrees: "))
    print("Sin =", math.sin(math.radians(angle)))

elif choice == 6:
    angle = float(input("Enter angle in degrees: "))
    print("Cos =", math.cos(math.radians(angle)))

elif choice == 7:
    angle = float(input("Enter angle in degrees: "))
    print("Tan =", math.tan(math.radians(angle)))

elif choice == 8:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("GCD =", math.gcd(a, b))

elif choice == 9:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("LCM =", math.lcm(a, b))

else:
    print("Invalid Choice")
```
