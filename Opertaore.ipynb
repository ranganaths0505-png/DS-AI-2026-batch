```python
#Combine comparison and logical operators to check if a year is a leap year (divisible by 4 and not divisible by 100, OR divisible by 400). Let year = 2024.

year=2024
is_leap_year = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

print("Is leap year:", is_leap_year)
```

```python
year = 2024

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

```python
#Perform 5 arithmetic assignments (+=, -=, *=, /=, %=) on a starting variable value 10 and print each step.

x = 10
print("Initial value:", x)

x += 5
print("After x += 5 :", x)

x -= 3
print("After x -= 3 :", x)

x *= 2
print("After x *= 2 :", x)

x /= 4
print("After x /= 4 :", x)

x %= 3
print("After x %= 3 :", x)
```
