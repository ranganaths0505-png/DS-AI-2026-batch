```python
#Convert a string '45' to an integer. Print the type before and after.
number_str = "45"

print(type(number_str))  # Before conversion

number_int = int(number_str)

print(type(number_int))  # After conversion
```

```python
#Convert a float 99.9 to an integer. Observe what happens to the decimal part.
price = 99.9

print(price)
print(type(price))

price_int = int(price)

print(price_int)
print(type(price_int))
```

```python
#Convert a boolean True to an integer and print it. What is the value?
is_active = True

result = int(is_active)

print(result)
print(type(result))
```
